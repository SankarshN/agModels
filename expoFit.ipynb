```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('barleyCropData.csv')
df = df.drop(['Unit for yield'], 1)
df = df.drop(['Unit for price'], 1)

df_price = df.drop(['Price'], 1)
x = np.array(df_price['Year']-1865)
y = np.array(df_price['Yield'])

df_yield = df.drop(['Yield'], 1)
x_2 = np.array(df_yield['Year']-1865)
y_2 = np.array(df_yield['Price'])

```

```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
#popt = np.array([0, 0, 0])
def func_exp(x, a, b, c):
        #c = 0
        return a * np.exp(b * x) + c

def exponential_regression (x_data, y_data):
    global popt
    popt, pcov = curve_fit(func_exp, x_data, y_data, p0 = (-1, 0.01, 1))
    print(popt)
    puntos = plt.plot(x_data + 1865, y_data + 1865, 'x', color='xkcd:maroon', label = "data")
    curva_regresion = plt.plot(x_data + 1865, func_exp(x_data, *popt) + 1865, color='xkcd:teal', label = "fit: {:.3f}, {:.3f}, {:.3f}".format(*popt))
    plt.legend()
    plt.show()
    return func_exp(x_data, *popt)
```
